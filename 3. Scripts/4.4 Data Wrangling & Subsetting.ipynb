#### - Transposing data

```python
# Import the “departments.csv” data set in to my notebook and save it into df_dep variable
df_dep = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'departments.csv'), index_col = False)
```

```python
# Check the first rows of the dataset we imported
# It is showing the whole dataset because it has only on row but many columns
df_dep.head()
```

```python
# Transpose the dataset by using .T which is capital T
# i.e. swich dataset from columns to rows and rows to columns
df_dep.T
```

```python
# Save it into a new variable (df_dep_t) after we transposed the dataset
df_dep_t = df_dep.T
```

```python
# Test the result of the new variable (df_dep_t)
df_dep_t
```

```python
# we can see the result above that index of the transposed dataset ws messed up
# We used a function .reset_index() to reset and organize the index of the dataset
df_dep_t.reset_index()
```

```python
# we are trying to organise the new dataset, as we can see the present header (index,0) is unclear and useless.
# We can address the new header by using a function .iloc(). and save it into a new variable (new_header)
# Now our new header starts at 1st row (index[0]) on the dataset
new_header = df_dep_t.iloc[0]
```

```python
# Test result of the new header
new_header
```

```python
# After having the new header, we need to have our data body too, so it can be completed and organised dataset as we want it to be
# By adressing index from the dataset df_dep_t, we can get any rows we want in our new organised dataset.
# [1:] tells the system to pick the data starting from 1st row and so on [:] until the last row
# Save it into the new variable (df_dep_t_new)
df_dep_t_new = df_dep_t[1:]
```

```python
#Test the new variable (df_dep_t_new)
df_dep_t_new
```

```python
# So far, we've got header part and body part
# We need to assemble them by using .columns attribute 
# To let the system know that data in the new header variable will be the column of data in df_dep_t_new variable
df_dep_t_new.columns = new_header
```

```python
# Test the result after assempling then together
df_dep_t_new
```
