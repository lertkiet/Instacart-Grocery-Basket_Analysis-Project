#### If else-statement

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
## Applying if-statement into the function we created 
df['price_range'] = df.apply(price_label, axis = 1)
```

#### 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Create a new list to save the result after running the for-loop according to the task's assignment
result2 = []
for value in ords_prods_merge['orders_day_of_week']:
  if value in (0,1):
    result2.append('Busiest days')
  elif value in (3,4):
    result2.append('Slowest days')
  else:
    result2.append('Regularly busy')
```

Explaination: 
I decided to use if-statement with for loop. From the counts of each value from 'orders_day_of_week' column as seen above, 
Saturday (0) and Sunday (1) are the busiest days.
Tuesday (3) and Wednesday (4) are the slowest days.
Other days (2,5,6) are Regularly busy days.

```python
# Result variable we created is a list. we need to save it into a dataframe in order to be able to execute any commands in Python
# Save the list into ords_prods_merge dataframe. By coding below, we are creating a new column ('Busiest days')
ords_prods_merge['Busiest days'] = result2
```

```python
# Use .head() function to check the result. We can see the last column ('Busiest days') was created
ords_prods_merge.head()
```

#### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Used .value.counts() of 'order_hour_of_day' column to see frequency of each hour 
ords_prods_merge['order_hour_of_day'].value_counts()
```

Split one third of values from 'order_hour_of_day' column into each group.

Hour: 10, 11, 14, 15, 13, 12, 16, 9 = Most orders

Hour: 17, 8, 18, 19, 20, 7, 21, 22  = Average orders

Hour: 23, 6, 0, 1, 5, 2, 4, 3       = Fewest orders

```python
# Create a new list to save the result after running the for-loop according to the task's assignment
result3 = []
for value in ords_prods_merge['order_hour_of_day']:
  if value in (10, 11, 14, 15, 13, 12, 16, 9):
    result3.append('Most orders')
  elif value in (17, 8, 18, 19, 20, 7, 21, 22):
    result3.append('Average orders')
  else:
    result3.append('Fewest orders')
```

```python
# Create a new column containing the labels called 'busiest_period_of_day.'
ords_prods_merge['busiest_period_of_day'] = result3
```
