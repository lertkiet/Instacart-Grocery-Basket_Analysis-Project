#### 5. Determine a suitable way to combine the orders_products_combined dataframe with your products data set. 
#### Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

```python
# Import products_checked file from 'Prepared data' folder
products_checked = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'))
```

```python
products_checked.head()
```

We are going to join 2 datsframe, I decided to drop the flag column from 'orders_products_combined' dataframe because it is not necessary anymore and, we will create another flag column after merging 2 new dataframe, this might confuse us.

```python
# drop the existing flag column
orders_products_combined = orders_products_combined.drop(columns = ['_merge'])
orders_products_combined
```

Both dataframes have product_id column in common. I decided to do inner join (default).

```python
# Merge 2 dataframes by using product_id column as a match
df_merged = orders_products_combined.merge(products_checked, on = ['product_id'], indicator = True)
df_merged
```

#### 6. Confirm the results of the merge using the merge flag.

```python
# Check sum of each value from the flag column (_merge)
df_merged['_merge'].value_counts()
```

The result tells us that both dataframes have a full match which seems to make sense because we did inner join (only all values match from 2 dataframes).
Can we trust this result?
I decided to run another test by using argument (how = outer), to check the the result of sum counts.

```python
# Because this is just a test, I am using pd.merge(), so It will not overwrite the existung data
# I used .value_counts() function as well to count the sum of each values from (how = outer)
(pd.merge(orders_products_combined,products_checked, on = ['product_id'], indicator = True))['_merge'].value_counts()
```

The result matched when we used how = inner, which is the default. Both dataframes have a full match.
