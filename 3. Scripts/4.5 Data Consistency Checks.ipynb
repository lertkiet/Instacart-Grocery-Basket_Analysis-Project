```python
# Check the result of the column we created
df_test.head()
```

```python
# Use for-loop to find out if there are any mixed data, if there are (Boolean = True), print command will work
for col in df_test.columns.tolist():
  weird = (df_test[[col]].map(type) != df_test[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_test[weird]) > 0:
    print (col)
```

```python
# Check if there are any missing data in each column from df_prods dataframe
df_prods.isnull().sum()
```

```python
# Because null values are in product_name column, we will not be able to replace the values with the code below
# df['column with missings'].fillna(mean value, inplace=True) 
# We decided to show only rows that are not missing out into a subset dataframe
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
# Check the size of the new dataframe which have non-empty values
df_prods_clean.shape
```
