#### 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

#### Creating a loyalty flag column using the transform()

```python
# Aggregating data with transform() function
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# This command tells pandas not to assign any options regarding the maximum number of rows to display
pd.options.display.max_rows = None
```

```python
# Testing the result with transform()
ords_prods_merge[['user_id', 'order_number', 'max_order']].head(100)
```

#### Creating a Loyalty flag Label using the loc()

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']  <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking the frequency of the new “loyalty_flag” column using the value_counts() function
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Show top 100 rows by 'user_id', 'loyalty_flag', 'order_number' columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(100)
```

#### 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
#### - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#### - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# 1. Aggregating Data with transform()
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Testing the output of the 'spending_flag' column
ords_prods_merge.head()
```

```python
# 2. Deriving Columns with loc()
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10 , 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10 , 'spending_flag'] = 'High spender'
```

```python
# Check the frequency of the new “loyalty_flag” column using the value_counts() function
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Testing the output of spending_flag column
ords_prods_merge[['user_id', 'average_price', 'spending_flag']].head(100)
```

#### 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#### - If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### - If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### - If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# 1. Aggregating Data with transform()
ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['total_days_since_last_order'].transform(np.median)
```

```python
# Testing the new column created
ords_prods_merge.head()
```

```python
# 2. Deriving Columns with loc()
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] <= 20) & (ords_prods_merge['median_days'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the frequency of the new “frenquency_flag” column using the value_counts() function
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking relevant columns of updated dataframe
ords_prods_merge[['user_id', 'total_days_since_last_order', 'median_days', 'frequency_flag']].head(100)
```
